Using the titanic dataset, build a model to predict if a passenger will survive or not based on age, gender...

```python
# Read and access data
import pandas as pd 

import os
os.chdir('E:\SkillEnable Class\Python\Datasets')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
# Selecting the variables
titanic2 = titanic[['pclass', 'survived', 'sex', 'age']]
titanic2.head()
```

```python
titanic2.isnull().sum()
```

```python
# Cleaning the data
titanic3 = titanic2.dropna()
```

```python
titanic3.head()
```

```python
# Creating dummies for 'sex' column
titanic4 = pd.get_dummies(titanic3, columns=['sex'])
```

```python
titanic4.head()
```

```python
Y = titanic4[['survived']]
```

```python
X = titanic4[['pclass', 'age', 'sex_female']]

# Splitting the data into test and train data.
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.85, random_state = 1)

len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
# Fitting the model into train data
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()

model_lr = LR.fit(X_train, Y_train)

model_lr
```

```python
Y_test
```

```python
# Testing the data
Y_test['Pred_survived'] = model_lr.predict(X_test)
```

```python
pd.crosstab(index = Y_test['survived'], columns = Y_test['Pred_survived'], margins = True)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test[['survived']], Y_test[['Pred_survived']])
accuracy
```
